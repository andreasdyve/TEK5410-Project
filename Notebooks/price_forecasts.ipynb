```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Script for calculating annuitised costs for PV, Wind, Gas, Gas with CCS and Carbon Tax 2020-2050

### Annuitised cost formula


$$
a = \frac{I_0}{ \sum_{t} \frac{1}{(1+r)^t} }
$$

## Gas CCGT + CCS

```python
# https://www.sciencedirect.com/science/article/pii/S2211467X18300634?via%3Dihub#tbl14

low = (588/857) # 2050/2020
high = (2113/2394) # 2050/2020
avg_decrease = (low+high)/2

# https://www.globalccsinstitute.com/archive/hub/publications/17011/costs-co2-capture-transport-and-storage.pdf
# page 42

cost_2020 = 1511
cost_2050 = cost_2020 * avg_decrease

years = [2020,2025,2030,2035,2040,2045,2050]
costs_CCGTCCS = np.linspace(cost_2020,cost_2050,len(years))
costs_CCGTCCS = costs_CCGTCCS*0.88 # Convert to EURO

annuitised_cost_CCGTCCS = np.zeros(7)
for i in range(7):
    denom = 0
    for j in range(1,31):
        denom += 1/((1+0.05)**j)
    annuitised_cost_CCGTCCS[i] = 1000*costs_CCGTCCS[i]/(denom)

fig,ax = plt.subplots(ncols=2,figsize=(9,4))
fig.tight_layout(pad=5.0)

ax[0].plot(years, costs_CCGTCCS, color='cadetblue')
ax[0].scatter(years, costs_CCGTCCS, color='red')
ax[0].set_xlabel('year')
ax[0].set_ylabel('Investment cost [EUR/kW]')
ax[0].set_title('CCGT + CCS Investment Cost')

ax[1].plot(years, annuitised_cost_CCGTCCS, color='cadetblue')
ax[1].scatter(years, annuitised_cost_CCGTCCS, color='red')
ax[1].set_xlabel('year')
ax[1].set_ylabel('Annuitised cost [EUR]')
ax[1].set_title('CCGT + CCS Annuitised Costs')
plt.show()

print(np.round(annuitised_cost, decimals=0))
```
